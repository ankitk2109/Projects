## I. None Feature Selection

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import cross_val_score
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.metrics import accuracy_score
from sklearn.feature_selection import SelectKBest, mutual_info_classif
from mlxtend.feature_selection import SequentialFeatureSelector as SFS
from mlxtend.plotting import plot_sequential_feature_selection as plot_sfs
```

```python
#Reading Training and Testing Files in DataFrames
train_df =  pd.read_csv('SPECTF_train.csv')
test_df = pd.read_csv('SPECTF_test.csv')

# Fetching all columns from training dataframe except the target column(DIAGNOSIS) to X_train and target to y_train.
X_train = train_df.iloc[:,1:]
y_train = train_df.iloc[:,0]

# Fetching all columns from testing dataframe except the target column(DIAGNOSIS) to X_test and target to y_test.
X_test = test_df.iloc[:,1:]
y_test = test_df.iloc[:,0]
```

### Gradient Boosting
Gradient Boosting builds an additive model in a forward stage-wise fashion; it allows for the optimization of arbitrary differentiable loss functions. In each stage n_classes_ regression trees are fit on the negative gradient of the binomial or multinomial deviance loss function


```python
#Creating a Gradient Boosting Classifier(gbc), loss = deviance loss function to be optimized. ‘deviance’ refers to 
#deviance (= logistic regression) for classification with probabilistic outputs.
gbc = GradientBoostingClassifier(loss = 'deviance', random_state=0)

#Fitting gbc with train data only.
gbc.fit(X_train,y_train)

#Testing the classifier on test data only and storing the value in y_pred
y_pred = gbc.predict(X_test)

#Finding the accuracy by comparing the y_pred and y_test.
acc_hold = accuracy_score(y_test,y_pred)
```

```python
#Applying Cross Validation on training data with cv=5. X_train contain all columns except the target column. y_train contains the target column.
scores = cross_val_score(gbc, X_train, y_train, cv=5)

#Taking average of all cross val scores.
acc_cross_val = scores.mean()

#Creating a dataframe of both the accuracies so that it would be easy to plot the bar graph.
df_None = pd.DataFrame([[acc_hold, acc_cross_val]],columns = ['Test','Train'], index=['None'])
print(df_None)
```

```python
#Creating a bar chart for comparing the accuracies from hold-out and cross validation
df_None.plot(kind='bar',figsize=(6,4),grid=True)
plt.ylabel('Accuracy')
plt.xlabel('All feautures')
plt.title('Accuracy Plot b/w Test and Train data')
```

## Feature Selection

In Feature selection we need to find the best subset of all available features, which contains the smallest number of features which contribute most to the accuracy. Discard the remaining, unimportant features.

**Feature Selection Strategies**
1. **Filter**: During Pre-processing step it ranks the features independently of the choice of classifier that will be subsequently applied. It returns a score value which is predictivness of the features. Example: Information Gain, Gini index.
2. **Wrapper**: The classifier is “wrapped” in the feature selection mechanism. Feature subsets are evaluated directly based on their performance when used with that specific classifier. A key aspect of wrappers is the search strategy which generates the candidate feature subsets for evaluation. One of the feature search stratergy is Sequential search. Two type of sequential search are given below:
    - **Forward Sequential Selection**: In this we start with an empty subset. Find the most informative feature and add it to the subset. Then repeat this process until there is no improvement by adding features.
    - **Backward Elimination**: Start with the complete set of features. Then remove the least informative feature and repeat until there is no improvement by dropping features.

## II. Information Gain

```python
#Mutual information (MI) of two random variables is a measure of the mutual dependence between the two variables. 
#More specifically, it quantifies the "amount of information" (in units such as shannons, commonly called bits) obtained 
#about one random variable through observing the other random variable
mi = {}
i_scores = mutual_info_classif(X_train, y_train)
#print(i_scores) #Information gain on all 44 features 

for i,j in zip(X_train.columns,i_scores): #Mapping column names to each i_scores values.
    mi[i]=j

#Creating a DataFrame for all the features and there i_score.
df = pd.DataFrame.from_dict(mi,orient='index',columns=['I-Gain'])

#Sorting the DataFrame based on I-Gain values in decending order.
df.sort_values(by=['I-Gain'],ascending=False,inplace=True)
```

```python
#Plotting Information Gain Scores vs Features on a bar chart.
n = len(df.index)
rr = range(1,n)
fig, ax = plt.subplots(figsize=(6,4), dpi=100)
ax.bar(df.index, df["I-Gain"], label='I-Gain',width=.35)
ax.set_xticklabels(list(df.index), rotation = 90)
ax.set_title('I-Gain score for each feature')
ax.set_xlabel('Features')
ax.set_ylabel('I-Gain')
ax.legend()
plt.show()
```

### Selecting *k* Best Features

We rank the features using information gain (well mutual information) and select the _k_ best to build a classifier.  
We iterate through increasing values of *k*.  
`SelectKBest` is a _transform_ that transforms the training data.

    - Out of 44 features it will first identify the feature with the max i_score.
    - Now it will keep on adding the features in decreasing order of i_score
    - In each iteration we are also fitting the model on X_train and y_train.
    - appending each accuracy in a list.

```python
acc_scores = []
for kk in range(1, X_train.shape[1]+1):
    #Fitting X_train and y_train in each iteration with varying k.
    FS_trans = SelectKBest(mutual_info_classif, k=kk).fit(X_train, y_train)
    
    #Transforming the X_train and X_test.
    X_tR_new = FS_trans.transform(X_train)
    X_tS_new = FS_trans.transform(X_test)
    
    #Fitting the model on new X_tR_new and y_train.
    seg_gbc = gbc.fit(X_tR_new, y_train)
    y_dash = seg_gbc.predict(X_tS_new)
    
    #Checking the accuracy of y_test and y_dash.
    acc = accuracy_score(y_test, y_dash)
    acc_scores.append(acc)

#Adding the accuracy column to the dataframe.
df['Accuracy'] = acc_scores
df.head(10)
```

```python
#Plotting the Accuracy vs I-Gain score on the graph to identify the best number of features required for training the classifier.
n = len(df.index)
rr = range(1,n)
fig, ax = plt.subplots(figsize=(6,4), dpi=100)
ax2 = ax.twinx()
ax.bar(df.index, df["I-Gain"], label='I-Gain',width=.35)
ax2.plot(df.index, df["Accuracy"], color='red', label='Accuracy')
ax.set_xticklabels(list(df.index), rotation = 90)
ax.set_xlabel('Features')
ax.set_ylabel('I-Gain')
ax2.set_ylabel('Accuracy')
ax.legend()

plt.show()
```

```python
#Pandas dataframe.idxmax() function returns index of first occurrence of maximum over requested axis.
selected_features_list = list(df[:df['Accuracy'].idxmax()].index.values)  #selected_features_list contains the list of named features that till the maximum accuracy.

#Fetching the above selected features from training and testing sets and storing them into new variable.
X_train_InfoGain = X_train[selected_features_list]
X_test_InfoGain = X_test[selected_features_list]

#Applying the gradient boosting classifier on the newly created X_train_InfoGain and y_train.
gbc.fit(X_train_InfoGain,y_train)

#Using gradient classifier to predict the test set.  
y_pred = gbc.predict(X_test_InfoGain)

#Calculating the accuracy score.
acc_hold_InfoGain = accuracy_score(y_test,y_pred)
print('Hold-out accuracy using I-Gain:',acc_hold_InfoGain)
```

```python
#Cross-Validation using selected features based on k-best features.
scores = cross_val_score(gbc, X_train_InfoGain, y_train, cv=5)
print('Each fold score:',scores)

acc_cross_val_InfoGain = scores.mean()
print ("Cross-validated mean scores:", acc_cross_val_InfoGain)

#Creating a dataframe of both the accuracies to plot the bar graph.
df_InfoGain = pd.DataFrame([[acc_hold_InfoGain, acc_cross_val_InfoGain]],columns = ['Test','Train'], index=['IG'])
print(df_InfoGain)
```

```python
#Creating a bar chart for comparing the accuracies from hold-out and cross validation using selected features.
df_InfoGain.plot(kind='bar',figsize=(6,5),grid = True)
plt.ylabel('Accuracy')
plt.xlabel('k-Best feautures')
plt.title('Accuracy Plot b/w Test and Train data')
```

## Wrapper

__scikit learn does not support Wrapper feature selection so we use MLxtend.__

 - Wrapper approache are based on greedy search algorithms as they compare all possible combinations and pick the combination that produces the best outcome for a given machine learning algorithm.
 
 - Step Forward Feature Selection: In the first phase, the classifier's performance is evaluated in accordance with each feature. Of all the features, the features that performs the best is selected.

 - In next step, the first feature is then tested in combination with all other features. The resultant features that produce the best performance are selected. 

#### Parameters:
   1. **k_features** : Specifies the number of features to select.
   2. **forward** : If set to True, performs step forward feature selection otherwise backward. 
   3. **verbose** : If parameter is used for logging the progress of the feature selector.
   4. **scoring** : Parameter defines the performance evaluation criteria.
   5. **cv** : Refers to cross-validation folds.
   6. **n_jobs** : If set to -1 then we run the cross-validation on all our available CPU cores.

## III. Wrapper-Forward

```python
#Creating a gradient Boosting classifier with random state set to zero.
gbc = GradientBoostingClassifier(loss = 'deviance', random_state=0)

#Selecting all columns from X_train and storing it into feature_names for further processing.
feature_names = X_train.columns

#Creating a sequential forward selector with all features.
sfs_forward = SFS(gbc, 
                  k_features=44, 
                  forward=True, 
                  floating=False, 
                  verbose=1,
                  scoring='accuracy',
                  cv=10, n_jobs = -1)

#Fitting the selector on training data with all features.
sfs_forward = sfs_forward.fit(X_train, y_train, custom_feature_names=feature_names)
```

```python
#Plotting sequential feature selection to find the highest accuracy feature.
fig1 = plot_sfs(sfs_forward.get_metric_dict(), #Returns a dictionary of feature_idx, cv_scores, avg_score, ci_bound, std_dev, std_err.
                ylabel='Accuracy',
                kind='std_dev', figsize=(8,6))

plt.ylim([0.5, 1])
plt.title('Sequential Forward Selection (w. StdDev)')
plt.grid()
plt.show()
print(sfs_forward.k_feature_names_)
```

```python
#Finding the heighest k value from the above graph.
df_res = pd.DataFrame(pd.DataFrame(sfs_forward.get_metric_dict()).loc['avg_score'].astype('float64'))
number_features = df_res.idxmax()
heighest_score_num = int(number_features[0])

# Top 'k' features:
sfs_forward.subsets_[heighest_score_num]['feature_names']

#Setting k_features = heighest_score_num and recalculating the sequential forward selector.
feature_names = X_train.columns
sfs_forward = SFS(gbc, 
                  k_features=heighest_score_num, 
                  forward=True, 
                  floating=False, 
                  verbose=1,
                  scoring='accuracy',
                  cv=10, n_jobs = -1)

sfs_forward = sfs_forward.fit(X_train, y_train, custom_feature_names=feature_names)
```

```python
#Plotting sequential feature selection graph with top k features.
fig1 = plot_sfs(sfs_forward.get_metric_dict(), #Returns a dictionary of feature_idx, cv_scores, avg_score, ci_bound, std_dev, std_err.
                ylabel='Accuracy',
                kind='std_dev', figsize=(8,6))

plt.ylim([0.5, 1])
plt.title('Sequential Forward Selection (w. StdDev)')
plt.grid()
plt.show()
print(sfs_forward.k_feature_names_)
```

```python
#sfs_forward.subsets_ #Subsets in each iteration

#Cross Val accuracy for selected features
print('Cross-Val Score:',sfs_forward.k_score_)

#Fetching k best features from train and test set and putting them into new variables for further training the classifier. 
selected_train_feature = X_train[list(sfs_forward.k_feature_names_)]
selected_test_features = X_test[list(sfs_forward.k_feature_names_)]

#Fitting gbc classifier on X & y Features.
gbc.fit(selected_train_feature,y_train)

#Predicting the y_pred_test_forward with selected feature test set.
y_pred_test_forward = gbc.predict(selected_test_features)
#y_pred_test_forward

#Hold-out accuracy.
acc_holdout_forward = accuracy_score(y_pred_test_forward,y_test)
#acc_holdout_forward

#Creating a dataframe of both the accuracies to plot the bar graph.
df_forward = pd.DataFrame([[acc_holdout_forward, sfs_forward.k_score_]],columns = ['Test','Train'], index=['Wrapper-F'])
#df_forward

#Creating a bar chart for comparing the accuracies from hold-out and cross validation using wrapper-forward selection.
df_forward.plot(kind='bar',figsize=(6,5),grid = True)
plt.ylabel('Accuracy')
plt.xlabel('k-Best feautures')
plt.title('Sequential Forward Selection')
```

## IV. Wrapper-B

```python
#Creating a gradient Boosting classifier with random state set to zero.
gbc = GradientBoostingClassifier(loss = 'deviance', random_state=0)

#Selecting all columns from X_train and storing it into feature_names for further processing.
feature_names = X_train.columns

#Creating a sequential backward elemination with all features.
sfs_backward = SFS(gbc, 
                  k_features = 1, 
                  forward = False, #Makes this a backward elemination stratergy.
                  floating = False, 
                  verbose = 1,
                  scoring = 'accuracy',
                  cv=10, 
                  n_jobs = -1)

#Fitting the selector on training data with all features.
sfs_backward = sfs_backward.fit(X_train, y_train, custom_feature_names=feature_names)
```

```python
#Plotting all features to find the highest accuracy feature.
fig2 = plot_sfs(sfs_backward.get_metric_dict(), #Returns a dictionary of feature_idx, cv_scores, avg_score, ci_bound, std_dev, std_err.
                ylabel='Accuracy',
                kind='std_dev', figsize=(8,6))

plt.ylim([0.5, 1])
plt.title('Sequential backward Selection (w. StdDev)')
plt.gca().invert_xaxis()
plt.grid()
plt.show()
print(sfs_backward.k_feature_names_)
```

```python
#Finding the heighest k value from the above graph.
df_res = pd.DataFrame(pd.DataFrame(sfs_backward.get_metric_dict()).loc['avg_score'].astype('float64'))
number_features = df_res.idxmax()
heighest_score_num = int(number_features[0])

# Top 'k' features:
sfs_backward.subsets_[heighest_score_num]['feature_names']

gbc = GradientBoostingClassifier(loss = 'deviance', random_state=0)
feature_names = X_train.columns

#Setting k_features = heighest_score_num and recalculating the backward elemination selector.
sfs_backward = SFS(gbc, 
                  k_features= heighest_score_num, 
                  forward=False, 
                  floating=False, 
                  verbose=1,
                  scoring='accuracy',
                  cv=10, n_jobs = -1)

sfs_backward = sfs_backward.fit(X_train, y_train, custom_feature_names=feature_names)
```

```python
#Plotting graph with top k features.
fig2 = plot_sfs(sfs_backward.get_metric_dict(), #Returns a dictionary of feature_idx, cv_scores, avg_score, ci_bound, std_dev, std_err.
                ylabel='Accuracy',
                kind='std_dev', figsize=(8,6))

plt.ylim([0.5, 1])
plt.title('Sequential backward Selection (w. StdDev)')
plt.gca().invert_xaxis()
plt.grid()
plt.show()
print(sfs_backward.k_feature_names_)
```

```python
#sfs_backward.subsets_ #Subsets in each iteration

#Cross Val accuracy for selected features
#print('',sfs_backward.k_score_)

#Hold-Out Stratergy on selected features
selected_train_feature_backward = X_train[list(sfs_backward.k_feature_names_)]
selected_test_features_backward = X_test[list(sfs_backward.k_feature_names_)]

gbc.fit(selected_train_feature_backward,y_train)

y_pred_test_backward = gbc.predict(selected_test_features_backward)
#y_pred_test_forward

acc_holdout_backward = accuracy_score(y_pred_test_backward,y_test)

df_backward = pd.DataFrame([[acc_holdout_backward, sfs_backward.k_score_]],columns = ['Test','Train'], index=['Wrapper-B'])

#Creating a bar chart for comparing the accuracies from hold-out and cross validation using wrapper-backward selection.
df_backward.plot(kind='bar',figsize=(6,5),grid = True)
plt.ylabel('Accuracy')
plt.xlabel('k-Best feautures')
plt.title('Sequential backward Selection')
```

# Conclusion

**FEATURE SELECTION METHOD**

1. FILTER METHOD: The Information Gain filter is an agnostic method of pre-selection classifier that is independent of the method of machine learning. This is more effective computationally, but the choice of features that rely on the learning algorithm it avoids.

2. WRAPPER METHOD: Wrappers are input approaches that integrate the Machine Learning algorithm into the feature selection process, i.e. to assess the quality of a set of features, they depend on the output of a particular classifier. 


_For each function that can be added to or deleted from the feature subset, wrapper methods scan through the space of feature subsets and measure the approximate precision of a single learning algorithm._

_The accuracy using all 3 feature selection strategies Information-Gain, Wrapper-Forward(Sequential Forward Selection), Wrapper-Backward (Backward - elimination) are consistent._

However the filter method is unstable as compared to the Wrapper based methods because:
1. No Model Bias: Different features may suit different learning algorithms but filters do not take this into account.

2. No Feature Dependencies:
    - The features are considered individually from one another, and are not considered in context.
    - In some cases, a filter can pick two predictive yet correlated characteristics where one is appropriate.
    - In other instances, one function requires another to improve accuracy, but this can not be found by a filter.


_The Wrapper based methods considers bias into account and considers the features in context._
_The wrapper approach is more rigorous computationally, but takes into account the constraints of the subset of functions on the learning algorithm._

Wrapper based Sequential Forward Selection feature selection method is faster than backward elimination technique as it starts with small subsets, so requires less running time if stopped early.
   1. It starts from 1 feature and goes upto the maximum accuracy bearing features but it's accuracy is less than that of the backward elimination strategy as once it gets the maximum accuracy in between, it will stop considering the next features. 
   2. The Wrapper based Backward Elimination strategy is slow but accuracy is high as it starts from combination of all the features calculating the accuracy and removing the feature one by one till no feature left.

The divergence between the training and test data can be accompanied by two facts:
1. The size of the training data set is not even half the test data set that can result in overfitting.
2. Using features extracted from wrapper methods can lead to overfitting as wrapper methods already train with the features machine learning models and reduce the true learning capacity. But in most situations, the functionality of filter methods will not result in overfitting.

Based on the resultant graph, the accuracy of the test dataset is similar throughout the selection strategy but the accuracy of the training dataset in the case of Wrapper-based methods eventually increases.

Filter methods may not find the best subset of features in cases where insufficient data are available to model the statistical correlation of features, but wrapper methods can always provide the best subset of features due to their exhaustive existence.
